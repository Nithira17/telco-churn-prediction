```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('data/processed/telco_churn_missing_values_handled.csv')
df.head(5)
```

```python
count = ((df['PhoneService'] == 'No') & (df['MultipleLines'] != 'No phone service')).sum()
print(f"Count: {count}")
```

- To reduce the confusions both 'No ___ service' values are also converted to 'No'

```python
df['MultipleLines'] = df['MultipleLines'].replace({'No phone service': 'No'})
df['OnlineSecurity'] = df['OnlineSecurity'].replace({'No internet service': 'No'})
df['OnlineBackup'] = df['OnlineBackup'].replace({'No internet service': 'No'})
df['DeviceProtection'] = df['DeviceProtection'].replace({'No internet service': 'No'})
df['TechSupport'] = df['TechSupport'].replace({'No internet service': 'No'})
df['StreamingTV'] = df['StreamingTV'].replace({'No internet service': 'No'})
df['StreamingMovies'] = df['StreamingMovies'].replace({'No internet service': 'No'})

df.head(15)
```

- Created a new feature `CommunicationType` by combining `PhoneService` and `InternetService` to capture customer communication patterns and interaction effects.


```python
def comm_type(row):
    if row['PhoneService'] == 'No' and row['InternetService'] == 'No':
        return 'No Service'
    elif row['PhoneService'] == 'Yes' and row['InternetService'] == 'No':
        return 'Phone Only'
    elif row['PhoneService'] == 'No' and row['InternetService'] != 'No':
        return 'Internet Only'
    else:
        return 'Phone and Internet'

df['CommunicationType'] = df.apply(comm_type, axis=1)
df.head(15)
    
```

```python
cols = df.columns.tolist()
insert_at = cols.index('InternetService') + 1
cols.remove('CommunicationType')
cols.insert(insert_at, 'CommunicationType')
df = df[cols]

df.head(5)
```

- Mapped 'Yes' and 'No' into 1 & 0

```python
internet_services = [
    'OnlineSecurity', 'OnlineBackup',
    'DeviceProtection', 'TechSupport',
    'StreamingTV', 'StreamingMovies'
]

df['TotalInternetServices'] = df[internet_services].replace({'Yes': 1, 'No': 0}).sum(axis=1)
df.head(5)
```

```python
cols = df.columns.tolist()
insert_at = cols.index('CommunicationType') + 1
cols.remove('TotalInternetServices')
cols.insert(insert_at, 'TotalInternetServices')
df = df[cols]

df.head(5)
```

```python
df.to_csv('data/processed/telco_churn_features_changed.csv', index=False)
```
