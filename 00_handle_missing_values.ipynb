```python
df = pd.read_csv('data/raw/telco-customer-dataset.xls')
df.head(5)
```

```python
df.isnull().sum()
```

- No NaN values are in the dataset
- But need to observe if there are any dataentry in unusual format

```python
columns = ['customerID', 'gender', 'SeniorCitizen', 'Partner',
           'Dependents', 'tenure', 'PhoneService', 'MultipleLines',
           'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling',
           'PaymentMethod', 'MonthlyCharges', 'TotalCharges', 'Churn']

for column in columns:
    unique_values = df[column].unique()
    print(f'{column}: {unique_values}')
```

- The data in TotalCharges column are in String format. Need to convert into numerical format

```python
df_dropped = df.drop(columns=['customerID'])
df_dropped.head(5)
```

- TotalCharges column consists ' ' instead of '0'. Need to handle this here since it turns into NaN value when converting to numerical format

```python
df_dropped[df_dropped['TotalCharges']==' ']
```

```python
df_dropped['TotalCharges'] = df_dropped['TotalCharges'].replace(' ', '0')
```

```python
df_dropped[df_dropped['TotalCharges']==' ']
```

```python
df_dropped['TotalCharges'] = pd.to_numeric(df_dropped['TotalCharges'], errors='coerce') 
df_dropped.head(5)
```

```python
df[df['TotalCharges'].isna()]
```

```python
df_dropped.to_csv('data/processed/telco_churn_missing_values_handled.csv', index=False)
```
